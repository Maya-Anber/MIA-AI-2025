```python
def display_gear(num):

    # handle invalid inputs: num must be an integer from 0 to 8
    if not isinstance(num, int) or num < 0 or num > 8:
        print("Error: Gear number must be an integer from 0 to 8")
        return
    
    # Each digit is represented as a list of 5 strings, each 4 characters long
    segment_patterns = {
    0: [
        "####\n",
        "#  #\n",
        "#  #\n",
        "#  #\n",
        "####\n"
    ],
    1: [
        "   #\n",
        "   #\n",
        "   #\n",
        "   #\n",
        "   #\n"
    ],
    2: [
        "####\n",
        "   #\n",
        "####\n",
        "#   \n",
        "####\n"
    ],
    3: [
        "####\n",
        "   #\n",
        "####\n",
        "   #\n",
        "####\n"
    ],
    4: [
        "#  #\n",
        "#  #\n",
        "####\n",
        "   #\n",
        "   #\n"
    ],
    5: [
        "####\n",
        "#   \n",
        "####\n",
        "   #\n",
        "####\n"
    ],
    6: [
        "####\n",
        "#   \n",
        "####\n",
        "#  #\n",
        "####\n"
    ],
    7: [
        "####\n",
        "   #\n",
        "   #\n",
        "   #\n",
        "   #\n"
    ],
    8: [
        "####\n",
        "#  #\n",
        "####\n",
        "#  #\n",
        "####\n"
    ]
    }

    # Get the pattern for the given gear number
    pattern = segment_patterns[num]

    for row in pattern:
        print(row, end='')
```

```python
gear = input("Enter Gear (0-8): ")


if gear.isdigit():
    # Convert input to integer and display gear
    gear_number = int(gear)
    display_gear(gear_number)

else: 
     # handle invalid inputs
     print("Invalid input. Please enter a number from 0 to 8.")
```
